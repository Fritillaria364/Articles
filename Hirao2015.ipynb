# 語順の相関に基づく機械翻訳の自動評価法
平尾努  


## 0. あぶすと
機械翻訳の開発には自動評価法が必須  
しかし、これまでの自動評価法(Nグラムを用いたBLEU, 最大共通部分単語列の割合に基づくROUGE-Lなど)は、人間が高く評価するルールベース機械翻訳(RBMT)の訳を低く評価する  
これは, RBMTがターゲットと全く違う単語を用いることが原因である  
これら従来の自動評価法は訳に用いる単語が一致しない場合スコアが激減するが、人間が評価する場合はその限りではない  
また、SMTで英日日英翻訳を行うとAなのでBと訳すべき箇所をBなのでAと訳しがちである.本来このような訳には低いスコアが与えられるべきだがNグラムの一致割合で見るとそこまでスコアは下がらない  

本稿では、訳語の違いに寛大かつ、大局的な語順を考慮した自動評価法を提案する  
大局的な語順の考慮には順位相関係数で測定  
訳語の違いは単語適合率で測定し、パラメータを用いてその重みを調整できるように  



## 1. はじめに
現行の自動評価法はクソ  

BLEUのようなNグラムに着目すると、システム翻訳が部分的に参照翻訳のNグラムを保持しているだけで高いスコアが与えられる(意味が乖離していようとも)  
このような現象は、翻訳時に大きな語順の入れ替えを必要としない言語間→構文が似ている言語間ではほとんど起こらない.  
つまり言語によってスコアのでやすさが大きく異なってくるしクソ  

ROUGE-LやIMPACTのような最長部分単語列に着目すると、大局的な語順を評価できるのでNグラムよりは英日日英において良い評価が期待できる  
しかし, Nグラムと共通する問題として訳語の違いに敏感すぎる  

ので本稿では日英英日のような翻訳時に大きな語順の入れ替えを必要とする言語対を対象とした翻訳システムの自動評価法について提案する  
提案手法の特徴は文全体における大局的な語順に着目する点、参照翻訳とシステム翻訳間で一致しない単語を採点から外し、別途ペナルティとしてそれをどの程度重要視するかを調整できるようにすることで訳語の違いに対し寛大な評価を行う点にある  

具体的な手法としては、翻訳間の語順の近さを測るため両者に一致して出現する単語を同定したのち、それらの出現順序の近さを順位相関係数を用いて計算、これに重み付き単語正解率と短い翻訳に対するペナルティを乗じたものを最終スコアとする
