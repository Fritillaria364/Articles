# 確率的トピックモデルによる対人認知傾向と性格特性の関連付け
熊野 史朗ほか

## 0 - あらまし
発話する側(認知側にとって刺激となる行動を取る側)のモデル化は今までいろいろな研究でなされてきた  
しかし、刺激を受け取る認知側、つまり知覚者の特定の個人差をモデル化した例はすくないのでそれをやる  
対人認知傾向：他者の感情について、それを受け取る知覚者が肯定・否定といった感情の軸上でのバイアルを持つかどうかみたいな  
この**対人認知傾向**と**性別**、**性格特性**の3要因について、確率的トピックモデルを用いて関連付けた結果、妥当な関係性が見いだせた  

## 1 - 提案手法
### 1.1 提案モデル
提案モデルとして、多様な人の間に少数の典型的人格があると仮定する  
各典型的人格は固有の対人認知傾向・性別・性格特性の3種の属性を確率的に持つ  
個々の知覚者の性質はこれら典型的人格の混合により記述されるものとする  
提案(生成)モデルは以下のプロセスを仮定する  
はじめに、対象知覚者が持つ典型的人格の混合率が事前分布$\pi$により決定される  
混合率が定まれば、対象者の刺激に対する認知・性別・性格特性がそれぞれ独立にK個の混合分布からのランダム抽出により決定する  
各属性は以下のような分布に従っており、$\Theta= \{ \pi_{k},\alpha_{k},\beta_{k},\mu_{k},\sigma_{k} \}^{K}_{k=1}$がパラメタとして設定される   

|属性|分布|
|:-----:|:-----:|
|対人認知傾向|多項分布  |
|性別|二項分布|
|性格特性|正規分布|

このモデルを用い、性別が$y_{j}$、性格特性が$s_{j}$である人物jが状態eと認知する確率を求めることを考える  

### 1.2 学習と推定
学習では対人認知傾向・性別・性格特性、つまり$\{ \alpha_{j},y_{j},s_{j} \}$が揃った学習データを用いてモデルパラメタ$\Theta$を学習する  
推定では学習した$\Theta$、対象人物jの性別yj及び性格特性sjを既知として、対人認知傾向$\alpha_{j}$を推定する  



## 2 - 実験
50人用意して実験を行い、対人認知タスクと性格特性を調査した  
それらのデータを用い、モデルパラメタの学習と対人認知傾向の推定を行った  
典型的人格として、実験対象を対人認知傾向を基準にグループ分割した結果を用いた  


## 3 - モデルパラメタの学習結果
実験を元に学習したパラメタの値を見ると、妥当な結果が得られた  
対人認知傾向において共感に偏った典型的人格は、性格特性の尺度でも同様に共感性のスコアが高いという結果が出ている  


## 4 - 対人認知傾向の推定評価
全被験者の認知傾向の平均を比較対象とし、推定した認知傾向の精度を評価した  
その結果、全被験者を平均して、比較法より有意な結果が見られた  
特に極端な認知傾向のあるグループについては大幅な精度向上が見られたため、平均的な認知傾向を持つ人物に対しても推定は有用であると思われる
