# 話の好感度に基づく自然言語発話からの情緒生起手法

## 1. 前文
　より快適なコミュニケーションを実現するためには, ユーザの感情を考慮した対話システムの構築が求められる.(特に自由対話)  
 情緒：ユーザの発話内容を, ユーザ自身が持つ嗜好情報に基づいて判断し, その結果感じた快/不快  
 気分：過去に生起された情緒の影響が累積したもの  
 本論では, ユーザに生起している情緒を計算するために発話事象に含まれる話の好感度と事象の格フレーム表現の意味構造に着目する  
 全12種の事象タイプに対し8種の情緒計算式を作成, 各語に対するユーザの好感度を代入することで発話事象に生起している情緒を計算する  
 

## 2. 感情のとらえかた
### 2.1 エージェントモデル
対話処理能力をもったエージェントのモデルを想定  
基本的な値泡処理の流れを以下のように策定  

1. ユーザ発話の解析(文解析)
1. 返答内容の決定(対話プランニング)
1. 発話文生成(文生成)

さらに, モデルは自己モデル・ユーザモデル、現在状況を保持する内界  
知識・エピソードといった長期記憶を保管する記憶管理領域を持つ  
外界からユーザの発話文が入力されると, 文解析により発話意図と発話内容の解析が行われる  
対話プランニング領域ではユーザの発話意図・対話履歴・現在の感情状態に基づき何を応答するか(what-to-say)と状況を加味した上でどのように返答するか(how-to-say)を決定する  
文生成領域では決定した返答内容に基づきユーザモデルや対話履歴を考慮して具体的な発話文字列を生成する  

これらの対話処理に, 様々な発話内容から生起するユーザの感情を計算する感情計算領域を追加する  
本研究で対象となるのはユーザの発話が表す事象であるため、おおよその状況は述語や格要素によって既に示されている  
そこで, 述語や格要素などの表層構造を入力とする単純な情緒計算手法を提示する  



# 3. 情緒生起

### 3.1 情緒計算プロセス
快/不快の判別・強度計算の2段階  
- 事象の快/不快判別  
事象内の格要素に対しユーザが持つ好感度を使用  
ex. ｢私が太郎に勝つ｣ → 私/太郎/勝つ  
これらの語の好感度を事象のタイプに応じた情緒計算式に代入し, 快不快を判別する  

### 3.2 好感度  
情緒計算式に用いられる語の好感度は, ユーザの好悪の度合いにより決定し, [-1.0.1.0]の実数値で表される  
好感度にはあらかじめ与えられるもの。状況に応じ自動更新が行われるものの二種がある  
情緒計算時に初期好感度データにない語が現れた場合, 概念データベースを検索し対象となる語の上位概念の好感度を適用する  
検索した結果, 不明であった語にはニュートラルの好感度を与える  

語に対する好感度の値は, 既知の語, また新しく出現した語に関して, ユーザの発話内容や発話時の表出感情から推測することでこれらの値を学習し更新することが可能である  
好感度の値更新は以下の4手法により履行される  

- ユーザ自身による好悪の表明を検出する  
ユーザが好き/嫌いという感情を明らかに表出させた際, その対象に正/負の好感度を与える(値は適当).  
- 好感度変化条件を用いる方法
人間や事物が好かれる/嫌われる状況をルールとして与えておくことで, 対話によって得られた状況に応じてそれらの好感度を変化させる  
このルールは質問紙形式であらかじめ調査することで得られる  
- 不快な事象内に現れるものの好感度を下げる
ある事象が不快であると認識した場合, その事象に含まれる語の好感度をわずかに下げる  
- 情緒計算式と現在の感情から逆算する  
ユーザの感情表出をフィードバックし, それをもとに情緒計算式を逆算することで好感度の値を求める  
ex. ユーザが｢私は〜をもらった｣と発話し, 状態が快であればユーザはもらった対象物を好きであると判断する  

これらの好感度学習の手法はユーザの個人情報に大きく依存するため, ユーザの嗜好の特徴を質問紙調査により認識し, それに応じた好感度の変化を行う.

### 3.4 情緒計算式
情緒計算式を算出する上で, まず快の感情が生起される条件を定める  

1. 好き/嫌いなものが利益/不利益となるような行動を取る  
1. 好き/嫌いなものの状況が好転/悪化する
1. 好き/嫌いなものが良い/悪い評価を受ける
1. 好き/嫌いなものが好き/嫌いな属性を持つ

不快の場合はこの逆となる  

情緒計算式で用いる$f_i(i\in{S,O,OF,OT,OM,OS,OC,P})$(P:事象そのもの)は各格要素に対する好感度を表す  

#### s1 事象タイプ
表1に示すそれぞれの格フレームに対する情緒計算式の考え方と具体的な計算式を与える  

Ⅰ: 主体Sがそれ自身に影響を及ぼすような事象Vを行うことを表す  
快生起条件1より, 情緒値は$f_S \times f_P$で与えられる  

Ⅱ,Ⅲ: 主体Sの事象Vのに関する状態が出発点OFから目標OTに変化したことを表す  
出る,行くといった位置・量の変化を示す場合, その変化量により生起する情緒値が定まる  
よって, 目標OTと出発点OFの好感度の差分をとることにより現在の状態から過去から好転/悪化したかを判断する  
情緒値は$f_S\times(f_{OT}-f_{OF})$で与えられる(OT,OFが対話中で特に述べられていない限りは0を与える)  

しかし, 事象タイプⅢには憧れる・適するといった精神, 感覚の変化を表す語も存在する  
これらに関しては変化内容そのものが情緒の生起に影響するため, 出発点は0とし, 状況の好転/悪化は事象自身の好感度Pの正負により表す  
したがって情緒値は$f_S\times(f_{OT}-0)\times f_P$  
最終的にこれらの情緒計算式をまとめると, $f_S\times(f_{OT}-f_{OF})\times f_P$で表される  

Ⅳ：主体Sとその相互作用の相手OMが事象Vという関係にある/なることを表す  
これら三要素の総積が情緒値($f_S\times f_{OM}\times f_P$)となる  

Ⅴ：AがBに勝つ  
Aを主体, Bを拠り所とした際, 主体Sにとっての出来事Aが勝つと拠り所OSにとっての出来事Bが勝たれる(負ける？)が同時に生起している  
このとき, 視点が逆になることで事象の好感度も逆転してしまう  
そこで, Sの出来事からの生起情緒とOSの出来事の生起情緒の和を全体の情緒とする  
よって, 情緒値は$(f_S\times f_P)+(f_{OS}\times (-f_P)) = (f_S-f_{OS})\times f_P$

Ⅵ：主体Sの作用に注目する場合と, 客体の反応に注目する場合の2通りがある  
前者の場合, 主体Sが客体Oに対し事象Vという行動を取ることを表す  
形としては事象タイプⅠである主体Sが事象Vを行うと同様であるが, 事象Vの好感度を考えるにあたり何に対してという要素が大きな影響力をもつ  
そのため, 情緒値は$f_S\times(f_O\times f_P)$で考える

一方, 後者の場合は主体Sによって客体Oに事象Vがなされることを表す  
この場合, 客体Oが事象Vをされたということが基本となるため, 事象の好感度は〜されることに関する好感度を用いる  
情緒値は$f_O\times f_P$となる  

Ⅶ,Ⅷ：タイプⅡ,Ⅲにおける主体Sが客体Oに置き換わったやつ  
よって情緒値は$f_S$を$f_O$に置き換えた$f_O\times(f_{OT}-f_{OF})\times f_P$で表される  

Ⅸ：事象タイプⅣと同様, 客体Oとその相互作用の相手が主体Sにより事象Vという関係になることを表す  
よって, 情緒値は$f_O\times f_{OM}\times f_P$となる  
ただし, このタイプに属する相互作用事象には"置換"に関わるものがいくつかあるが, この事象はそれまでの対話履歴や既知の知識を検索し推論する必要がある. このため, 単一事象のみからの情緒生起を取り扱う本研究ではこのタイプについては情緒計算式を作成しないこととする  

Ⅹ：道具Iは情緒値の強度にのみ影響を及ぼすため, 事象の表す意味はタイプⅠと相違ない  
ただ, 情緒計算式は$f_S$を$f_O$に置き換えた式$f_O\times f_P$を用いる  

Ⅺ：客体Oが属性OCであることを表す  
情緒値は$f_O\times f_{OC}$になる  

#### s2 否定表現
"勝たない"や"強くない"など, 発話に否定の様相を持っている事象は, 事象に対する好感度$f_P$の正負を逆転させる  

#### s3 連体修飾構造
格要素に含まれる名詞は, 殆どの場合何らかの修飾構造が付与されている  
連体修飾構造には連体詞・名詞など句構造によって名詞を修飾する連体修飾句と, 文形式の構造によって名詞を修飾する連体修飾節がある  

- 連体修飾句  
修飾部に名詞・形容詞を含む場合, それらによって被修飾語に属性が与えられていると考える  
そのため, 修飾部に含まれる名詞・形容詞の好感度と被修飾語を掛けあわせた結果を句全体の好感度とする  
全体の好感度=修飾部×被修飾語  
こそあの3系列に関しては指示対象の好感度をそのまま修飾部の好感度とする  

- 連体修飾節  
被修飾語が指し示す対象が修飾節により限定されるか否かで計算式を変える  
内容節は被修飾語が特定概念を指し示しておらず, 修飾節の内容がそのまま被修飾語の概念となるため, 修飾節に情緒計算式を適用し得られた値を全体の好感度とする  
補足語修飾節のうち限定的修飾も被修飾語が固定の概念ではないため, 同様の処理を行う  
一方非限定的修飾の場合は被修飾語はある固有のものを指しているため, その後自体が固定された好感度を持っている  
そのため, 修飾説は情緒の強度を考慮する場合のみ用い, 情緒値の正負には影響しないものと考え全体の好感度は被修飾語の好感度をそのまま適用する  

### 3.5 情緒強度計算手法
前節では語の好感度を掛けあわせて快不快を判別する情緒計算式を示した  
しかし, これらの式は8種のパターンがあり、さらに二次式と三次式が混在している上, 好感度の絶対値が1以下であるために好感度を掛けあわせるほど平均的な値は小さくなる  
そこで, 本研究では情緒計算式の各項を直交ベクトル(たかだか3つ)と考えてこれらのベクトルからなる直方体の対角線の長さを生起情緒の強度とする  
また, 2個の項からなる計算式では3番目の項として0.5を与える  
ただし, 事象タイプⅩについては第三項として道具の格要素を考慮する
