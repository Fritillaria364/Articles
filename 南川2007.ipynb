# テキストマイニングによる個人ブログデータからの性格推定手法
南川 敦宣ら KDDI研究所

## 1. はじめに
本研究では、個人ブログで使用される語句からエゴグラムを推定することを考える  
テキストマイニングによりエゴグラム推定に有効な表現をブログから抽出し、ナイーブベイズ法により推定する  


## 2. エゴグラム
### 2-1 エゴグラム
自我状態の表現モデルの一種
CP(Critical Parent：批判的親 - 正義感・責任感・義務感/批判的・保守的)  
NP(Nurturing Parent: 養育的親 - 慈愛・寛容/過保護・過干渉)  
A(Adult: 大人 - 理性的・客観性/冷徹)
FC(Free Child: 自由な子供 - 自由奔放・創造的/自己中心的)
AC(Adapted Child: 従順な子供 - 従順・謙虚・協調性/反抗的・自虐的)

### 2-2 エゴグラム抽出
エゴグラムの抽出には以下の2手法がある  
- 観察による診断
- 質問紙法

## 3. ブログからのエゴグラム推定手法
ブログの文書分類問題としてみる  
執筆されたテキストデータを形態素解析し、得られた単語群をbag-of-wordsとしてみなしナイーブベイズ・SVM等の分類器を用いて分類を行う  
自我状態ごとに分類器を用意し、それぞれの分類器でその出力結果を分類クラスとして推定する  
bag-of-wordsから各分類器への入力ベクトルを生成する際の着目語(素性)の選択は、教師データにおける各単語の情報利得の大きさを基準に行う  
$$InfoGain(w) = H(C)-H(C|X)$$
$$H(C) = -\sum_{c\in C}P(c)\log P(c)$$
$$H(C|X) = -\sum_{x_w\in X}P(X=x_w)H(C|X=x_w)$$
w: 単語, $c\in C$: 分類結果となる各自我状態のスコア, $x_w$: 語wの出現頻度
提案手法における素性選択では、表現に関する品詞(形容詞・副詞),機能語に分類される品詞(接続詞・助動詞・助詞)、感情表現に使用される品詞(感嘆詞・記号・未知語)を対象とした  
分類器にはMNB(Multinomial Naive Bayes: 多項ナイーブベイズ)を使用した  



## 4. 評価実験
### 4-1 実験データ
実験データとして、551人のブログ筆者からブログデータを収集し、さらに執筆者にはTEG2を受診してもらった回答結果を使用した  
エゴグラム推定精度の評価は上記データを10分割交差検定によって行う  

### 4-2 評価シナリオ
素性数を50から200まで50刻みで変化させ、以下の4つの分類に関し評価を行った  

- (e1).各自我状態を2段階評価  
    全被験者の各自が状態の平均値を上回るか、下回るか
  
- (e2).各自我状態を3段階評価  
    TEG2の評価を(5,4) (3) (2,1)の3段階に変換

- (e3) 各自我状態を5段階評価  
    TEG2スコアをそのまま用いる

- (e4) スコア最大の自我状態を推定  
    各被験者の素性値と最大スコアの自我状態のペアを教師データとして使用

### 4-3 評価結果
ベースラインに全対象のうち最大頻度のクラスを推定結果とするやつ  
精度がベースラインより改善されたよ
